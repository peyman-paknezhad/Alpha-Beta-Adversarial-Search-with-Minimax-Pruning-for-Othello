```python
class othello():
    
    def __init__(self):
        self.board = [[0 for i in range(8)]for j in range(8)] # board
        self.board[3][3] , self.board[4][4] = -1 , -1 # set start state
        self.board[3][4] , self.board[4][3] = 1 , 1 #set start state
        self.n_white = 2 # number of whites
        self.n_black = 2 # nuumber of blacks
        self.white = set([(3,4) , (4,3)])
        self.black = set([(3,3) , (4,4)])
        
    def Board(self):# print the board
        print("\n_______________board_______________")
        c = 0 # row counter
        for i in self.board:
            print("\n")
            print ("row",c," :",end="")
            c += 1
            for j in i:
                if j == -1:
                    print(" "+"2",end="") # blacks show as 2
                else:    
                    print(" "+str(j),end="") 
        print("\n")           
    
            
    def is_terminal(self,player):
        if  len(self.white) + len(self.black) == 64:# there is no empty place
            return 1
        elif self.valid_moves(player) == -1: # no valid move for player
            player = self.take_turn(player) # change player to opponent
            if self.valid_moves(player) == -1:# no valid move for opponent
                return 1
        else:
            return -1

    def play(self,x,y,player):
        place = self.is_valid(x,y,player)# firs disc with same color in each col , row , dia
        if  place != -1:
                self.move(x,y,player)
                #row
                for i in range(y-place[0]-1): # berfor (x,y)
                    self.move( x,y-(i+1),player)
                for i in range(place[1]-y-1): #after (x,y)
                    self.move( x,y+(i+1),player)                    
                #col
                for i in range(x-place[2]-1):# berfor (x,y)
                    self.move( x-(i+1), y,player)
                for i in range(place[3]-x-1):#after (x,y)
                    self.move( x+(i+1), y,player)                    
                #n-dia
                for i in range(x-place[4]-1):# berfor (x,y)
                    self.move( x-(i+1), y-(i+1),player)  
                for i in range(place[5]-x-1):#after (x,y)
                    self.move( x+(i+1), y+(i+1),player)
                #p-dia
                for i in range(place[6]-x-1):# berfor (x,y)
                    self.move(x+(i+1), y-(i+1),player)
                for i in range(x-place[7]-1):#after (x,y)
                    self.move(x-(i+1), y+(i+1),player)  
                 

                            
    def move(self,x,y,player):
        if player == "w":
            if self.board[x][y] == -1:
                self.black.discard((x,y))
                self.n_black -= 1
            self.board[x][y] = 1
            self.white.add( (x,y) )
            self.n_white += 1
        else:
            if self.board[x][y] == 1:
                self.white.discard((x,y))
                self.n_white -= 1
            self.board[x][y] = -1
            self.black.add( (x,y) )
            self.n_black += 1
        
        
    def is_valid(self ,x,y,player):
        """
    return the place of first same color in row,col,dia
    if the move is not valid return -1
        """
        
        if self.board[x][y] != 0: # if it is not empty it's not valid
            return -1
        p = player
        
        
        counter = 0 # for count the number of discs that should change
        if p == 'w':
                #Row1 : check the row befor (x,y)
                counter = 0
                while(True):
                    if y - counter == 0: # it is on border of board
                        counter = 0
                        break
                    counter += 1
                    if y-counter > 0:# check the place is valid
                        if self.board[x][y-counter] == 1: # the first white disc
                            break
                        elif self.board[x][y-counter] == 0: # not valid to change in this direction
                            counter = 0
                            break
                        else: # it's black >>> keep going
                            continue
                    else: # not valid
                        counter = 0
                        break
                if counter > 1:# we should change and it's valid 
                    row1 = y - counter # col of first same color 
                else:
                    row1 = y  # there is no change          

                #Row2 : check the row after (x,y)
                counter = 0
                while(True):
                    if y + counter == 7: #it is on border of board
                        counter = 0
                        break
                    counter += 1
                    if y+ counter <7 :#check the place is valid
                        if self.board[x][y+counter] == 1:# the first white disc
                            break
                        elif self.board[x][y+counter] == 0:# not valid to change in this direction
                            counter = 0
                            break
                        else:# it's black >>> keep going
                            continue
                    else:
                        counter = 0
                        break
                if counter > 1: # we should change and it's valid 
                    row2 = y + counter # col of first same color 
                else:
                    row2 = y # there is no change   

                #col1 : check the col befor (x,y)
                counter = 0
                while(True):
                    if x - counter == 0: #it is on border of board
                        counter = 0
                        break
                    counter += 1
                    if self.board[x - counter][y] == 1:# the first white disc
                        break
                    elif self.board[x - counter][y] == 0:# not valid to change in this direction
                        counter = 0
                        break
                    else:# it's black >>> keep going
                        continue
                if counter > 1: # we should change and it's valid 
                    col1 = x - counter # row of first same color 
                else:
                    col1 = x # there is no change           
                # col 2 : check the col after (x,y)
                counter = 0
                while(True):
                    if x + counter == 7: #it is on border of board
                        counter = 0
                        break
                    counter += 1
                    if self.board[x + counter][y] == 1:# the first white disc
                        break
                    elif self.board[x + counter][y] == 0:# not valid to change in this direction
                        counter = 0
                        break
                    else:# it's black >>> keep going
                        continue
                if counter > 1: # we should change and it's valid 
                    col2 = x + counter # row of first same color 
                else:
                    col2 = x # there is no change  

                #n-dia1 : check the n_dia befor (x,y)
                counter = 0
                while(True):
                    if x - counter == 0 or y - counter == 0:#it is on border of board
                        counter = 0
                        break
                    counter += 1
                    if self.board[x - counter][y - counter] == 1:# the first white disc
                        break
                    elif self.board[x - counter][y - counter] == 0:# not valid to change in this direction
                        counter = 0
                        break
                    else:# it's black >>> keep going
                        continue
                if counter > 1: # we should change and it's valid 
                    n_dia1 = x - counter # row of first same color  
                else:
                    n_dia1 = x # there is no change              

                #n_dia2: check the n_dia after (x,y)
                counter = 0
                while(True):
                    if x + counter == 7 or y + counter == 7:#it is on border of board
                        counter = 0
                        break
                    counter += 1
                    if self.board[x + counter][y + counter] == 1:# the first white disc
                        break
                    elif self.board[x + counter][y + counter] == 0:# not valid to change in this direction
                        counter = 0
                        break
                    else:# it's black >>> keep going
                        continue
                if counter > 1: # we should change and it's valid 
                    n_dia2 = x + counter  # row of first same color  
                else:
                    n_dia2 = x # there is no change

                #p-dia1 :check the n_dia befor (x,y)
                counter = 0
                while(True):
                    if x + counter == 7 or y - counter == 0:#it is on border of board
                        counter = 0
                        break
                    counter += 1
                    if self.board[x + counter][y - counter] == 1:# the first white disc
                        break
                    elif self.board[x + counter][y - counter] == 0:# not valid to change in this direction
                        counter = 0
                        break
                    else:# it's black >>> keep going
                        continue
                if counter > 1: # we should change and it's valid 
                    p_dia1 = x + counter # row of first same color  
                else:
                    p_dia1 = x  # there is no change            

                #p_dia2 ::check the n_dia after (x,y)
                counter = 0
                while(True):
                    if x - counter == 0 or y + counter == 7:#it is on border of board
                        counter = 0
                        break
                    counter += 1
                    if self.board[x - counter][y + counter] == 1:# the first white disc
                        break
                    elif self.board[x - counter][y + counter] == 0:# not valid to change in this direction
                        counter = 0
                        break
                    else:# it's black >>> keep going
                        continue
                if counter > 1: # we should change and it's valid 
                    p_dia2 = x - counter # row of first same color 
                else:
                    p_dia2 = x   # there is no change                
        # for balck it's just like white
        else:#black

                #Row1
                counter = 0
                while(True):
                    if y - counter == 0:
                        counter = 0
                        break
                    counter += 1
                    if y-counter > 0:
                        if self.board[x][y-counter] == -1:
                            break
                        elif self.board[x][y-counter] == 0:
                            counter = 0
                            break
                        else:
                            continue
                    else:
                        counter = 0
                        break
                if counter > 1: 
                    row1 = y - counter 
                else:
                    row1 = y            

                #Row2
                counter = 0
                while(True):
                    if y + counter == 7:
                        counter = 0
                        break
                    counter += 1
                    if y+ counter <7 :
                        if self.board[x][y+counter] == -1:
                            break
                        elif self.board[x][y+counter] == 0:
                            counter = 0
                            break
                        else:
                            continue
                    else:
                        counter = 0
                        break
                if counter > 1: 
                    row2 = y + counter 
                else:
                    row2 = y

                #col1
                counter = 0
                while(True):
                    if x - counter == 0:
                        counter = 0
                        break
                    counter += 1
                    if self.board[x - counter][y] == -1:
                        break
                    elif self.board[x - counter][y] == 0:
                        counter = 0
                        break
                    else:
                        continue
                if counter > 1: 
                    col1 = x - counter 
                else:
                    col1 = x            
                # col 2
                counter = 0
                while(True):
                    if x + counter == 7:
                        counter = 0
                        break
                    counter += 1
                    if self.board[x + counter][y] == -1:
                        break
                    elif self.board[x + counter][y] == 0:
                        counter = 0
                        break
                    else:
                        continue
                if counter > 1: 
                    col2 = x + counter 
                else:
                    col2 = x 

                #n-dia1
                counter = 0
                while(True):
                    if x - counter == 0 or y - counter == 0:
                        counter = 0
                        break
                    counter += 1
                    if self.board[x - counter][y - counter] == -1:
                        break
                    elif self.board[x - counter][y - counter] == 0:
                        counter = 0
                        break
                    else:
                        continue
                if counter > 1: 
                    n_dia1 = x - counter 
                else:
                    n_dia1 = x             

                #n_dia2
                counter = 0
                while(True):
                    if x + counter == 7 or y + counter == 7:
                        counter = 0
                        break
                    counter += 1
                    if self.board[x + counter][y + counter] == -1:
                        break
                    elif self.board[x + counter][y + counter] == 0:
                        counter = 0
                        break
                    else:
                        continue
                if counter > 1: 
                    n_dia2 = x + counter 
                else:
                    n_dia2 = x

                #p-dia1
                counter = 0
                while(True):
                    if x + counter == 7 or y - counter == 0:
                        counter = 0
                        break
                    counter += 1
                    if self.board[x + counter][y - counter] == -1:
                        break
                    elif self.board[x + counter][y - counter] == 0:
                        counter = 0
                        break
                    else:
                        continue
                if counter > 1: 
                    
                    p_dia1 = x + counter 
                else:
                    p_dia1 = x             

                #p_dia2
                counter = 0
                while(True):
                    if x - counter == 0 or y + counter == 7:
                        counter = 0
                        break
                    counter += 1
                    if self.board[x - counter][y + counter] == -1:
                        break
                    elif self.board[x - counter][y + counter] == 0:
                        counter = 0
                        break
                    else:
                        continue
                if counter > 1: 
                    p_dia2 = x - counter 
                else:
                    p_dia2 = x               

            
        answer = [row1,row2,col1,col2,n_dia1,n_dia2,p_dia1,p_dia2] # positon of discs we should change discs between them
        
        if answer == [y,y,x,x,x,x,x,x]: # if there is no valid change it's not valid
            return -1
        
        return [row1,row2,col1,col2,n_dia1,n_dia2,p_dia1,p_dia2]
    
    
    
    def valid_moves(self,player): # retrun the valid moves
        """check all the board and check if it's valid to move for player or not
        """
        values = []
        for i in range(8):
            for j in range(8):
                if self.is_valid(i,j,player) != -1 :
                    values.append([i,j])
        if len(values) == 0:# if there is no valid move
            return -1
        else:
            return values  # a list of position in board      
                
    def save_state(self):# save the current state
        b = copy.deepcopy(self.black) # copy black dsics positions
        w = copy.deepcopy(self.white) # copy white discs positions
        return [b,w]

    def load_state(self,b,w): # load a state
        self.white = copy.deepcopy(w) # set whit discs
        self.black = copy.deepcopy(b) # set black discs
        self.n_white = len(self.white) # set number of whites
        self.n_black = len(self.black) # set number of blacks
        # set the board
        self.board = [[0 for i in range(8)]for j in range(8)] 
        for i in self.black:
            self.board[i[0]][i[1]] = -1
        for i in self.white:
            self.board[i[0]][i[1]] = 1            
          
    
    
    def value(self,state,a,b,depth,player): # minimax value function
        board = othello() # a new board
        board.load_state(state[0],state[1])# load the state
        if  board.is_terminal(player) == 1: # if it's a terminal state
            return board.n_black  # terminal value
        elif player == 'b' :#max agent
            return board.max_val(state,a,b,depth -1)
        else:#min agent
            return board.min_val(state,a,b,depth -1)
            
        
    def max_val(self,state,a,b,depth):
        board = othello()#a new board
        board.load_state(state[0],state[1])# load the state
        
        new = othello() #a new board
        
        if depth < 0: # depth limition
            return board.eval_function() # state value
        v = float('-inf')
        if board.valid_moves('b') == -1: # there is no valid move for black
            return new.min_val(state,a,b,depth-1)# white turn
        for i in board.valid_moves('b'): # for state successors
            board.play(i[0],i[1],'b') # change the state
            s = board.save_state() # saving the state
            temp = new.value(s,a,b,depth,'w') #val for new state
            if type(temp) == tuple: # return an action and eval-function and it's not terminal state
                temp = temp[0] # temp = eval(s)
            if temp > v: 
                v = temp # max v and new state
                action = i # set the action, max action 
            if v >= b: # pruning
                return v,i
            a = max(a,v) # set alpha
            board.load_state(state[0],state[1]) # reset the board
        return v , action     
        
    def min_val(self,state,a,b,depth):
        board = othello()#a new board
        board.load_state(state[0],state[1])# load the state
        
        new = othello()#a new board
        
        if depth  < 0:   # depth limition
            return board.eval_function() # state value
        v = float('+inf')
        if board.valid_moves('w') == -1:# there is no valid move for black
            return new.max_val(state,a,b,depth-1)# black turn
        for i in board.valid_moves('w'):# for state successors    
            board.play(i[0],i[1],'w') # change the state
            s = board.save_state()# save the state
            temp = new.value(s,a,b,depth,'b')#val for new state
            if type(temp) == tuple:# return an action and eval-function and it's not terminal state
                temp = temp[0]# temp = eval(s)
            if temp < v:
                v = temp #min v and new state
                action = i #set the action, min action 
            if v <= a:# pruning
                return v,i
            b = min(b,v) #set beta
            board.load_state(state[0],state[1]) # reset the board
        return v , action    
  

    def take_turn(self,player):  # change turn
        if player == 'b':
            player ='w'
            return player
        else:
            player ='b'
            return player
            
    def eval_function(self): # for the black discs
        n = 0
        player = 'b'
        x = -1
        for i in self.board:
            n += i.count(x) # number o black discs
        
        #number of board border black dsics
        n_b_b = 0
        n_b_b += self.board[0].count(x)
        n_b_b += self.board[7].count(x)
        for i in range(1,7):
            if self.board[i][0] == x:
                n_b_b += 1
            if self.board [i][7] == x:
                n_b_b += 1
        #numebr of corner black discs
        n_c = 0
        if self.board[0][0] == x:
            n_c += 1
        if self.board[0][7] == x:
            n_c += 1
        if self.board[7][0] == x:
            n_c += 1   
        if self.board[7][7] == x:
            n_c += 1            
        #number of valid moves for black
        valid_moves = self.valid_moves(player)
        if valid_moves != -1:    
            n_v_m = len(self.valid_moves(player))
        else:
            n_v_m = 0
        
        eval_f = n * 0.01  + n_b_b * 0.1 + n_c * 10 + n_v_m * 1
        return eval_f
    
    
    def n_empty(self):
        return 64 - (self.n_black + self.n_white)

    
    def game(self):
        player = 'b'# first player
        a = float('-inf') #set alpha
        b = float('inf') # set beta
        s = self.save_state() # save the state
        while(True):
            if self.is_terminal(player) == 1: # termina state == end game
                    self.Board() # print board
                    self.score() # print scores
                    print("**** GAME OVER ****")
                    break
            print("next move :\n")    
            e_func,action = self.value(s,a,b,DEPTH,player) # next move
            self.play(action[0],action[1],player) # play
            self.Board() # print board
            print("move = ",action ," state value = ",e_func ,"player = ",player)
            print("_________________________________")
            a = float('-inf') #set alpha
            b = float('inf') # set beta
            s = self.save_state() # set new state for minimax
            player = self.take_turn(player) # change the player
            
                
    def score(self):
        print(" white = ", self.n_white)
        print(" black = ", self.n_black)

```
